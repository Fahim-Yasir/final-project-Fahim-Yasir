## **Attack Classification**

```python
Train_attack = df_train.outcome.map(lambda a: 0 if a == 'normal' else 1)
Test_attack = df_test.outcome.map(lambda a: 0 if a == 'normal' else 1)

df_train['attack_state'] = Train_attack
df_test['attack_state'] = Test_attack
```

## **One-hot Encoding**

```python
df_train = pd.get_dummies(df_train,columns=['protocol_type','service','flag'],prefix="",prefix_sep="")
df_train
```

```python
df_test = pd.get_dummies(df_test, columns=['protocol_type', 'service', 'flag'], prefix="", prefix_sep="")
df_test = df_test.reindex(columns=df_train.columns, fill_value=0)
df_test
```

## **Data Split**

```python
X_train = df_train.drop(['outcome', 'level', 'attack_state'], axis=1)
Y_train = df_train['attack_state']
Xtrain,Xvalid,Ytrain,Yvalid = train_test_split(X_train, Y_train, test_size= 0.25 , random_state=42)
```

```python
Xtest = df_test.drop(['outcome', 'level', 'attack_state'], axis=1)
Ytest = df_test['attack_state']
```

## **Data Scalling**

```python
Ro_scaler = RobustScaler()
Xtrain = Ro_scaler.fit_transform(Xtrain)
Xvalid= Ro_scaler.transform(Xvalid)
Xtest = Ro_scaler.fit_transform(Xtest)
```

## **Logistic Regression**

```python
LR = LogisticRegression()
LR.fit(Xtrain, Ytrain)
LR.score(Xtrain,Ytrain), LR.score(Xvalid,Yvalid)
```

```python
RocCurveDisplay.from_estimator(LR,Xtest,Ytest)
```

## **Gaussian Naive Bayes**

```python
GNB = GaussianNB()
GNB.fit(Xtrain, Ytrain)
GNB.score(Xtrain,Ytrain), GNB.score(Xvalid,Yvalid)
```

```python
RocCurveDisplay.from_estimator(GNB,Xtest,Ytest)
```

## **K-Nearest Neighbors**

```python
n = range(1, 10, 2)
for n_neighbors in n:
  KNN = KNeighborsClassifier(n_neighbors=n_neighbors)
  KNN.fit(Xtrain,Ytrain)
  score = KNN.score(Xtrain,Ytrain), KNN.score(Xvalid,Yvalid)
  print(f"n_neighbors: {n_neighbors}, Score: {score}")
```

```python
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(Xtrain,Ytrain)
KNN.score(Xtrain,Ytrain), KNN.score(Xvalid,Yvalid)
```

```python
RocCurveDisplay.from_estimator(KNN,Xtest,Ytest)
```

## **Decison Tree with Grid Search**

```python
param_grid = {
    'max_depth': range(1, 10),
    'max_features': range(1, 10)
}

kfold = KFold(n_splits=10)

DT = DecisionTreeClassifier()
grid_search = GridSearchCV(estimator=DT, param_grid=param_grid, cv=kfold, scoring='roc_auc', verbose=1)

grid_search.fit(Xtrain,Ytrain)

print("Best parameters:", grid_search.best_params_)
```

```python
DT = DecisionTreeClassifier(max_depth=9,max_features=9)
DT.fit(Xtrain, Ytrain)
DT.score(Xtrain,Ytrain), DT.score(Xvalid,Yvalid)
```

```python
RocCurveDisplay.from_estimator(DT,Xtest,Ytest)
```

## **Gradient Boosting with Random Search**

```python
param_dist = {
    'n_estimators': randint(50, 150),
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': randint(1, 10)
}

kfold = KFold(n_splits=10)

GB = GradientBoostingClassifier()
random_search = RandomizedSearchCV(estimator=GB, param_distributions=param_dist, cv=kfold, scoring='roc_auc', verbose=1)

random_search.fit(Xtrain, Ytrain)

print("Best parameters:", random_search.best_params_)
```

```python
GB = GradientBoostingClassifier(max_depth=9,n_estimators=100,learning_rate = 0.05)
GB.fit(Xtrain, Ytrain)
GB.score(Xtrain,Ytrain), GB.score(Xvalid,Yvalid)
```

```python
RocCurveDisplay.from_estimator(GB,Xtest,Ytest)
```

## **PCA Analysis with Random Forest**

```python
Xcombined = np.concatenate([Xtrain, Xvalid, Xtest], axis=0)
pca = PCA(n_components=20)
pca = pca.fit(Xcombined)
Xreduced = pca.transform(Xcombined)
```

```python
Xtrain_reduced = Xreduced[:len(Xtrain)]
Xvalid_reduced = Xreduced[len(Xtrain):len(Xtrain) + len(Xvalid)]
Xtest_reduced = Xreduced[len(Xtrain) + len(Xvalid):]
```

```python
RF_original= RandomForestClassifier(max_depth=9)
RF_original.fit(Xtrain, Ytrain)
score_original = RF_original.score(Xtrain,Ytrain), RF_original.score(Xvalid,Yvalid)
RF_reduced= RandomForestClassifier(max_depth=9)
RF_reduced.fit(Xtrain_reduced, Ytrain)
score_PCA = RF_reduced.score(Xtrain_reduced,Ytrain), RF_reduced.score(Xvalid_reduced,Yvalid)
print(f"Original Data Score: {score_original}")
print(f"Reduced Data Score: {score_PCA}")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(13, 5))
RocCurveDisplay.from_estimator(RF_original,Xtest,Ytest,ax=ax[0])
ax[0].set_title('ROC Curve - Original Data')
RocCurveDisplay.from_estimator(RF_reduced,Xtest_reduced,Ytest,ax=ax[1])
ax[1].set_title('ROC Curve - Reduced Data')
```

## **Support Vector Machine with Different Kernel**

```python
Linear_SVC = svm.LinearSVC(C=1)
Linear_SVC.fit(Xtrain, Ytrain)
score_linear = Linear_SVC.score(Xtrain, Ytrain), Linear_SVC.score(Xvalid,Yvalid)
Poly_SVC = SVC(kernel='poly',C=1)
Poly_SVC.fit(Xtrain, Ytrain)
score_poly = Poly_SVC.score(Xtrain, Ytrain), Poly_SVC.score(Xvalid,Yvalid)
print(f"Linear SVM Score: {score_linear}")
print(f"Poly SVM Score: {score_poly}")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(13, 5))
RocCurveDisplay.from_estimator(Linear_SVC,Xtest,Ytest,ax=ax[0])
ax[0].set_title('ROC Curve - Linear SVM')
RocCurveDisplay.from_estimator(Poly_SVC,Xtest,Ytest,ax=ax[1])
ax[1].set_title('ROC Curve - Poly SVM')
```

```python
RBF_SVC = SVC(kernel='rbf',C=1)
RBF_SVC.fit(Xtrain, Ytrain)
score_gaussian = RBF_SVC.score(Xtrain, Ytrain), RBF_SVC.score(Xvalid,Yvalid)
Sigmoid_SVC = SVC(kernel='sigmoid',C=1)
Sigmoid_SVC.fit(Xtrain, Ytrain)
score_sigmoid = Sigmoid_SVC.score(Xtrain, Ytrain), Sigmoid_SVC.score(Xvalid,Yvalid)
print(f"RBF SVM Score: {score_gaussian}")
print(f"Sigmoid SVM Score: {score_sigmoid}")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(13, 5))
RocCurveDisplay.from_estimator(RBF_SVC,Xtest,Ytest,ax=ax[0])
ax[0].set_title('ROC Curve - RBF SVM')
RocCurveDisplay.from_estimator(Sigmoid_SVC,Xtest,Ytest,ax=ax[1])
ax[1].set_title('ROC Curve - Sigmoid SVM')
```

## **Ensemble with Voting Classifier**

```python
model1 = GaussianNB()
model2 = DecisionTreeClassifier(max_depth=9,max_features=9)
model3 = GradientBoostingClassifier(max_depth=9,n_estimators=100,learning_rate = 0.05)
model4 = RandomForestClassifier(max_depth=9)
ensemble = VotingClassifier(estimators=[('GNB', model1), ('DT', model2), ('GB', model3), ('RF', model4)], voting='soft')
ensemble.fit(Xtrain, Ytrain)
ensemble.score(Xtrain,Ytrain), ensemble.score(Xvalid,Yvalid)
```

```python
RocCurveDisplay.from_estimator(ensemble,Xtest,Ytest)
```

## **Neural Network**

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=64, activation='relu', input_shape=(Xtrain.shape[1:])),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(units=512, activation='relu'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(units=1, activation='sigmoid'),
])
model.compile(optimizer='adam', loss=tf.keras.losses.BinaryCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model.fit(Xtrain,Ytrain, validation_data=(Xvalid, Yvalid), epochs=10, verbose=1)
```

```python
predictions = model.predict(Xtest)
fpr, tpr, thresholds = roc_curve(Ytest, predictions)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='blue', lw=1, label='Neural Network (AUC = %0.2f)' % roc_auc)
plt.xlabel('False Positive Rate (Positive label: 1)')
plt.ylabel('True Positive Rate (Positive label: 1)')
plt.legend(loc="lower right")
```
